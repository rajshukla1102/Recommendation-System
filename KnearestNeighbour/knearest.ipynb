```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df_movie= pd.read_csv('movies.csv',usecols=['movieId','title'])
df_rating=pd.read_csv('ratings.csv',usecols=['userId', 'movieId', 'rating'])
```

```python
df_movie.head()
```

```python
df_rating.head()

```

```python
df=pd.merge(df_movie,df_rating,on='movieId')
```

```python
df.head()
```

```python
combine=df.dropna(axis=0,subset=['title'])
r_count = combine.groupby(by = ['title'],as_index=False)['rating'].count().rename(columns = {'rating': 'totalRatingCount'})
r_count.head()
```

```python
rating_totRating=combine.merge(r_count)
rating_totRating.head()
```

```python
r_count['totalRatingCount'].describe()
```

```python
popularity=50
pop_movie=rating_totRating.query('totalRatingCount>=@popularity')
pop_movie.head()
```

```python
df_pivot=pop_movie.pivot_table(index='title',columns='userId',values='rating').fillna(0)
df_pivot.head(10)
```

```python
# pop_movie1=pop_movie['title']=='10 Things I Hate About You (1999)'
# pop_movie[pop_movie1]['rating'].count()
# count() vs sum()
# count()-->It's a string function, which count the total number
# of individual category. and It only counts the number of non missing values
# where as sum() returns sum the number of unique category
```

```python
from scipy.sparse import csr_matrix
pop_movie_matrix=csr_matrix(df_pivot.values)

from sklearn.neighbors import NearestNeighbors

knn=NearestNeighbors(metric='cosine',algorithm='brute')
knn.fit(pop_movie_matrix)
```

```python
query_index=np.random.choice(df_pivot.shape[0])
print(query_index)
distances,indices=knn.kneighbors(df_pivot.iloc[query_index,:].values.reshape(1,-1),
                            n_neighbors=5)
# indices
# distances--->array([[1.11022302e-16, 5.18143727e-01, 5.87484041e-01, 5.96917195e-01,
#         5.98383464e-01]])


# distances.flatten()--->array([3.33066907e-16, 2.34651482e-01, 2.86554183e-01, 3.50092132e-01,
#        3.83449235e-01])
```

```python
# df_pivot.head()
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print(f"recommendation for {df_pivot.index[query_index]} ")
    else:
        print(pop_movie.index[indices.flatten()[i]])
```
